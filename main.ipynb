```python
#converts string literals to lists
import ast
ast.literal_eval(dataset.iloc[0].genres)
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L 
```

```python
dataset['genres'] = dataset['genres'].apply(convert)
```

```python
dataset.head()
```

```python
dataset['keywords'] = dataset['keywords'].apply(convert)
```

```python
dataset.head()
```

```python
dataset['cast'][0]
```

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if(counter != 3):
            L.append(i['name'])
            counter += 1
    return L 
```

```python
dataset['cast'] = dataset['cast'].apply(convert_cast)
```

```python
dataset['cast'][0]
```

```python
dataset.head()
```

```python
dataset['crew'][0]
```

```python
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            L.append(i['name'])
    return L 
```

```python
dataset['crew'] = dataset['crew'].apply(convert_crew)
```

```python
dataset.head()
```

```python
dataset['overview'] = dataset['overview'].apply(lambda x:x.split())
```

```python
dataset.head()
```

```python
def collapse(obj):
    L = []
    for i in obj:
        L.append(i.replace(" ",""))
    return L
```

```python
dataset['cast'] = dataset['cast'].apply(collapse)
dataset['crew'] = dataset['crew'].apply(collapse)
dataset['genres'] = dataset['genres'].apply(collapse)
dataset['keywords'] = dataset['keywords'].apply(collapse)
```

```python
dataset.head()
```

```python
dataset['tags'] = dataset['genres'] + dataset['cast'] + dataset['crew'] + dataset['keywords'] + dataset['overview']
```

```python
dataset.head()
```

```python
new = dataset[['id', 'title', 'tags']]
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x:" ".join(x))
```
